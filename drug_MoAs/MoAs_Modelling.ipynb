```python
import numpy as np
import os as os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.metrics import roc_auc_score
warnings.filterwarnings("ignore")


#setting up dataset paths
base_path = r"C:\Users\Mehrab\test\Kaggle_project_data\drug_MoAs"
save_path = r"C:\Users\Mehrab\test\Kaggle_projects\drug_MoAs"
train_features_path = os.path.join(base_path, 'train_features.csv')
train_targets_path = os.path.join(base_path, 'train_targets_scored.csv')
train_targets_nonscored_path = os.path.join(base_path, 'train_targets_nonscored.csv')

#reading in datasets
features = pd.read_csv(train_features_path)
targets = pd.read_csv(train_targets_path)


# filter warnings
import warnings
warnings.filterwarnings('ignore')

r_seed = 8

#! pip install iterative-stratification
```

<h2>Task description:</h2>
The goal is to use the features to predict the mechanism of action of a series of different drugs, encoded in target.

There is a large number of predictive, numeric features (872) and a large number of possible labels (206). Each row can have multiple labels, ie. this is a multi-label classification problem with a large number of sparsely labelled target classes.

Key findings from EDA:
1. Different target classes are present in very different frequencies - need stratified sampling. The problem defined on Kaggle requires us to minimize log-loss as the model performance metric, this is good because accuracy would be a bad metric.
2. There is an anonymized label column that needs to be dropped, ordinal dosage and time columns that should be encoded and a categorical treatment v/s control column that should also be encoded. 
3. There are no missing values in the dataset.
4. Cell-viability features are highly skewed and need to be power-transformed.
5. Cell-viability features are correlated and could be dimensionally reduced.
6. Target classes are relatively un-correlated, so an ensemble, one v/s all model should not pay much penalty by ignoring target class correlations.

<h2>Data pre-processing</h2>
<h3>Splitting data into training and validation sets.</h3>

```python
#splitting data with stratification to preserve ratios of different target classes across training and validation sets
#Doing only one split for now
splitter_obj = MultilabelStratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = r_seed)
for train_idx, valid_idx in splitter_obj.split(features, targets):
    X_train, y_train = features.iloc[train_idx], targets.iloc[train_idx]
    X_valid, y_valid = features.iloc[valid_idx], targets.iloc[valid_idx]
   
X_train_orig, y_train_orig, X_valid_orig, y_valid_orig = X_train, y_train, X_valid, y_valid
X_train.head()
```

<br/><br/><br/><br/>
<h2>Pre-processing data based on parameters learned from training set.</h2>
<h3>Down-sampling cell-viability feature columns with a PCA transformation.</h3>

```python
cxx_cols =  [i for i in X_train.columns if i.startswith('c-')]  #list of c-xx (cell viability) feature cols
gxx_cols =  [i for i in X_train.columns if i.startswith('g-')]  #list of c-xx (cell viability) feature cols

#PCA projection to 50 features to downsample cell-viability features
n_PCs = 60    #determined manually after trying a few numbers 
pca_obj = PCA(n_components = n_PCs)
pca_obj.fit(X_train[cxx_cols])    #PCA object fitted on training data
PC_col_names = ['c-PC' + str(num) for num in range(n_PCs)]
PCs_train = PCs_train = pd.DataFrame((pca_obj.transform(X_train[cxx_cols])), columns = PC_col_names)
PCs_valid = PCs_valid = pd.DataFrame((pca_obj.transform(X_valid[cxx_cols])), columns = PC_col_names)

#dropping original cxx columns and concatenating PCs
X_train.reset_index(drop=True, inplace=True)
X_valid.reset_index(drop = True, inplace = True)
PCs_train.reset_index(drop=True, inplace=True)
PCs_valid.reset_index(drop=True, inplace=True)
#dropping
c_indices = np.where(X_train.columns.str.startswith('c-'))
X_train = X_train.drop(X_train.iloc[:, c_indices[0]], axis = 1)
X_valid = X_valid.drop(X_valid.iloc[:, c_indices[0]], axis = 1)

#concatenating PC dims
X_train = pd.concat([X_train, PCs_train], axis = 1)
X_valid = pd.concat([X_valid, PCs_valid], axis = 1)

X_train_orig, X_valid_orig = X_train, X_valid
```

<h3>Pre-processing step 2</h3>
Encoding ordinal columns and robust-scaling them. Then Power transforming cxx-PC columns to reduce skewness inherited from cxx columns.

```python
#dropping anonymized sig_id column
X_train = X_train_orig.drop(['sig_id'], axis = 1)
y_train = y_train_orig.drop(['sig_id'], axis = 1)
X_valid = X_valid_orig.drop(['sig_id'], axis = 1)
y_valid = y_valid_orig.drop(['sig_id'], axis = 1)
all_col_names = X_train.columns

#Specifying order for ordinal encoding
ord_cols = ['cp_type', 'cp_time', 'cp_dose']
type_cats = ['ctl_vehicle', 'trt_cp'];
time_cats = [24, 48, 72]
dose_cats = ['D1', 'D2']


#Setting up a column transformer that does the rest of the pre-processing steps
ord_pipe = Pipeline(steps = [
    ('1_ord_enc', OrdinalEncoder(categories = [type_cats, time_cats, dose_cats])), 
    ('2_rob_sc', RobustScaler())
    ])

prepr_trf = ColumnTransformer([
    ('1_ord_pipe', ord_pipe, ord_cols), 
    ('2_pwr_tr', PowerTransformer(), PC_col_names)
    ], remainder = 'passthrough')

#transforming data
X_train = prepr_trf.fit_transform(X_train)
X_valid = prepr_trf.transform(X_valid)

#converting col transformer output arrays to DFs
X_train = pd.DataFrame(X_train, columns = all_col_names)
X_valid = pd.DataFrame(X_valid, columns = all_col_names)

X_train.head()
```

Using the Binary relevance ensemble method with an array of n_classes logistic regression classifiers. Since each classifier is fitted independently, a nice way to quickly reduce the time taken to quickly try out multiple models is to just fit the first 50 classes (ot of 205) and find the best model type for this target subset and then scale up.

<h3>Conclusions:</h3>
The random forest classifier works better than the Logistic regression classifier. The ensemble, Binary relevance method tends to have not too bad log loss scores, but the individual feature classifiers tend to over or under predict their class occurances, depending on the class, even with training set data. This might be because class output data might is still imbalanced, in spite of using stratified sampling. One way to fix this might be to use SMOTE - to synthetically over-sample under-represented classes by linear-interpolation.

It is also possible that the models start performing better with some hyperparameter tuning. But since the number of training samples is ~ 20000 and the number of target classes is > 200, each round of model training takes too long on a single CPU for this to be practical. One approach to work around this would be to use GPUs to run parallel versions of the logistic regression and random forest classifiers from the cuML library.
